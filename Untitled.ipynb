```python
import csv
```

```python
with open("dependencies.tsv") as f:
    dtsv = []
    reader = csv.reader(f, delimiter='\t')
    for r in reader:
        dtsv.append(r)
```

```python
dtsv
```

```python
dl['projects'][0]
```

```python
dl_dict = {i['name']: i for i in dl['projects']}
dl_dict
```

```python
for i in dtsv:
    name = i[0]
    sha = i[2]
    if name not in dl_dict:
        print("not in found in lock file ---> ", name)
        continue
    print(
        "|{}|{}|{}|".format(sha, sha == dl_dict[name]['revision'], name)
    )
```
